```python
# Sample dataset
data = [1, 2, 1, 3, 2, 4, 2, 3, 1, 5]

# Create a DataFrame from the dataset
df = pd.DataFrame(data, columns=['Value'])

# Calculate frequency distribution
frequency_table = df['Value'].value_counts().reset_index()
frequency_table.columns = ['Value', 'Frequency']

# Display the frequency distribution table
print("Frequency Distribution Table:")
print(frequency_table)

# Visualize the distribution using Matplotlib
plt.bar(frequency_table['Value'], frequency_table['Frequency'], color='skyblue')
plt.title('Frequency Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data for temperature and rainfall for each month of the year
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
temperature = [20, 22, 25, 28, 30, 32, 32, 30, 28, 25, 22, 20]
rainfall = [10, 8, 15, 20, 25, 30, 35, 30, 25, 20, 15, 10]

# Scatter plot - Temperature vs. Rainfall
plt.figure(figsize=(10, 5))
plt.scatter(temperature, rainfall, color='blue')
plt.title('Temperature vs. Rainfall for Each Month')
plt.xlabel('Temperature (°C)')
plt.ylabel('Rainfall (mm)')
plt.grid(True)
plt.show()

# Histogram - Temperature distribution
plt.figure(figsize=(10, 5))
plt.hist(temperature, bins=5, color='orange', edgecolor='black')
plt.title('Temperature Distribution')
plt.xlabel('Temperature (°C)')
plt.ylabel('Frequency')
plt.grid(axis='y')
plt.show()

# Histogram - Rainfall distribution
plt.figure(figsize=(10, 5))
plt.hist(rainfall, bins=5, color='green', edgecolor='black')
plt.title('Rainfall Distribution')
plt.xlabel('Rainfall (mm)')
plt.ylabel('Frequency')
plt.grid(axis='y')
plt.show()

# Box plot - Temperature distribution across seasons
seasons = ['Winter', 'Spring', 'Summer', 'Autumn']
seasonal_temperatures = [[20, 22, 25], [28, 30, 32], [32, 30, 28], [25, 22, 20]]
plt.figure(figsize=(10, 5))
plt.boxplot(seasonal_temperatures, labels=seasons, patch_artist=True, notch=True, vert=0)
plt.title('Temperature Distribution Across Seasons')
plt.xlabel('Temperature (°C)')
plt.ylabel('Season')
plt.grid(axis='x')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Sample dataset with independent variable x and dependent variable y
x = np.array([1, 2, 3, 4, 5])  # Independent variable
y = np.array([2, 4, 5, 4, 5])  # Dependent variable

# Calculate the slope (m) and intercept (b) of the regression line
m, b = np.polyfit(x, y, 1)

# Visualize the data and regression line
plt.scatter(x, y, color='blue', label='Data')  # Scatter plot of the data
plt.plot(x, m*x + b, color='red', label='Regression Line')  # Regression line
plt.xlabel('Independent Variable (x)')
plt.ylabel('Dependent Variable (y)')
plt.title('Simple Linear Regression')
plt.legend()
plt.grid(True)
plt.show()

# Print the slope and intercept of the regression line
print("Slope (m):", m)
print("Intercept (b):", b)
```

```python
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt

# Sample dataset with independent variable x and dependent variable y
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 5, 4, 5])

# Fit a simple linear regression model using statsmodels
X = sm.add_constant(x)  # Add a constant term for the intercept
model = sm.OLS(y, X).fit()

# Summary of the regression model
print(model.summary())

# Visualize the results
# Scatter plot of actual vs predicted values
predicted_y = model.predict(X)
plt.scatter(x, y, color='blue', label='Actual')  # Scatter plot of the actual values
plt.plot(x, predicted_y, color='red', label='Predicted')  # Regression line
plt.title('Simple Linear Regression')
plt.xlabel('Independent Variable (x)')
plt.ylabel('Dependent Variable (y)')
plt.legend()
plt.show()

# Residual plot
residuals = model.resid
plt.scatter(x, residuals, color='green')
plt.title('Residual Plot')
plt.xlabel('Independent Variable (x)')
plt.ylabel('Residuals')
plt.show()
```

```python
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt

# Sample dataset with independent variables x1, x2, x3 and dependent variable y
x1 = np.array([1, 2, 3, 4, 5])
x2 = np.array([2, 3, 4, 5, 6])
x3 = np.array([3, 4, 5, 6, 7])
y = np.array([10, 12, 14, 16, 18])

# Fit a multilinear regression model using statsmodels
X = sm.add_constant(np.column_stack((x1, x2, x3)))  # Add a constant term for the intercept
model = sm.OLS(y, X).fit()

# Summary of the regression model
print(model.summary())

# Visualize the results
# Scatter plot of actual vs predicted values
predicted_y = model.predict(X)
plt.scatter(y, predicted_y)
plt.plot(y, y, color='red')  # Plotting y = x line
plt.title('Actual vs Predicted Values')
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.grid(True)
plt.show()

# Residual plot
residuals = model.resid
plt.scatter(predicted_y, residuals)
plt.title('Residual Plot')
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.grid(True)
plt.show()
```

```python
# Sample dataset with independent variable x and dependent variable y
x <- c(1, 2, 3, 4, 5)
y <- c(2, 4, 5, 4, 5)

# Fit a simple linear regression model
model <- lm(y ~ x)

# Summary of the regression model
summary(model)

# Visualize the results using ggplot
library(ggplot2)

# Create a data frame with the independent variable and the predicted values
df <- data.frame(x = x, y = y, predicted_y = predict(model))

# Create a scatter plot of the actual values and the predicted values
ggplot(df, aes(x = x, y = y)) +
  geom_point(color = "blue") +
  geom_line(aes(x = x, y = predicted_y), color = "red") +
  labs(title = "Simple Linear Regression", x = "Independent Variable (x)", y = "Dependent Variable (y)")

# Create a residual plot
ggplot(df, aes(x = x, y = resid(model))) +
  geom_point(color = "green") +
  labs(title = "Residual Plot", x = "Independent Variable (x)", y = "Residuals")
```
